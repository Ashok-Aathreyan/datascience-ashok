```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sal = pd.read_csv('/content/Salaries.csv')
```

```python
print(sal.info())
```

```python
highest_paid = sal.loc[sal['TotalPayBenefits'].idxmax(), 'EmployeeName']
print(highest_paid)
```

```python
lowest_paid = sal.loc[sal['TotalPayBenefits'].idxmin(), 'EmployeeName']
print(lowest_paid)
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='Year', y='TotalPayBenefits', data=sal, marker='o')
plt.title('Trend of TotalPayBenefits Over Years')
plt.xlabel('Year')
plt.ylabel('TotalPayBenefits')
plt.grid(True)
plt.show()
```
